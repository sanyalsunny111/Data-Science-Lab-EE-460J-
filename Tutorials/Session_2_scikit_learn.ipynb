### Example 1: digit classification
##### Loading an example dataset

```python
from sklearn import datasets
digits=datasets.load_digits()
print(digits.data)
print(digits.target)
```

```python
print(digits.data.shape)
```

##### Learning and predicting

```python
from sklearn import svm
clf = svm.SVC(gamma=0.001, C=100.)
clf.fit(digits.data[:-1], digits.target[:-1])
```

```python
clf.predict(digits.data[-1:])
```

Let's check whether this is a good prediction by visualizing the input data with matplotlib:

##### Install matplotlib:
<code>conda install matplotlib</code>

```python
import matplotlib.pyplot as plt
plt.imshow(digits.data[-1:].reshape(8,8), cmap = plt.cm.gray)
print("ground truth is:", digits.target[-1])
```

### Example 2: classification of iris dataset with decision tree
##### Let's try the example from the lecture: 

```python
from sklearn import datasets
iris=datasets.load_iris()
print(iris.data)
print(iris.target)
```

```python
# shuffle the iris dataset
import numpy as np
shuffler = np.random.permutation(len(iris.data))
data_shuffled = iris.data[shuffler]
target_shuffled = iris.target[shuffler]
print(data_shuffled)
print(target_shuffled)
```

```python
# Generating our training and testing dataset
train_data = data_shuffled[:100]
train_target = target_shuffled[:100]
test_data = data_shuffled[100:]
test_target = target_shuffled[100:]
```

```python
# Now let's train a decision tree
from sklearn import tree
decision_tree = tree.DecisionTreeClassifier(max_depth=2)
decision_tree = decision_tree.fit(train_data, train_target)
tree.plot_tree(decision_tree) 
```

```python
# Let's see how well we did. What's the training accuracy?
train_score = decision_tree.score(train_data,train_target)
print("Train score: %.4f" % train_score)
```

```python
# Not bad!! How about the accuracy on the testing samples -- what we really care about.
test_score = decision_tree.score(test_data,test_target)
print("Train score: %.4f" % test_score)
```
